### inference

```python
my_model = YOLO("plp_detection_weights/best.pt")
```

```python
results = my_model("my_dataset/My_collected_car_images", save=True, show_boxes=False)
```

```python
import os
# Get the list of all files and directories
path = "my_dataset/My_collected_car_images"
dir_list = os.listdir(path)
print("Files and directories in '", path, "' :")
# prints all files
print(dir_list)
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
print("The bounding boxes in [xmin, ymin, xmax, ymax] format:")
i = 1
for file_name in dir_list:
    results = my_model(f"my_dataset/My_collected_car_images/{file_name}", save=True)
    img = cv2.imread(f"runs/detect/predict/{file_name}")
    img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
    for result in results:
        j = 1
        print(f"\nimage {i}:")
        boxes = result.boxes.xyxy.cpu().numpy()
        for box in boxes:
            print(f"Box:{box}")
            ymin = int(box[1])+5
            ymax = int(box[3])-5
            xmin = int(box[0])
            xmax = int(box[2])
            crop_img = img[ymin:ymax, xmin:xmax]
            # plt.imshow(crop_img)
            crop_img = cv2.cvtColor(crop_img , cv2.COLOR_BGR2RGB)
            cv2.imwrite(f"my_dataset/cropped_license_plates/lp_{i}_{j}.jpg", crop_img)
            j+=1

        i+=1
```
